**distribution of legit and fraudlent transaction**

The term "Class" usually refers to the target variable or label that indicates whether a transaction is fraudulent or not. It is a binary classification problem, where:

**Class 0:** Represents a legitimate (non-fraudulent) transaction.

**Class 1:** Represents a fraudulent transaction.





```python
df['Class'].value_counts()
```

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

Under Sampling:
Building dataset contatining similar distribution of normal transaction and fraudulent transaction.

fraudulent transaction -> 142

```python
legit_sample = legit.sample(n=142)
```

Concatenating two dataframes

```python
new_df = pd.concat([legit_sample, fraud],axis=0)
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting data into Features and Targets

```python
X = new_df.drop(columns='Class',axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training logisticregression model using training data
model.fit(X_train, Y_train)
```

Model Evaluation:
ACCURACY:

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```
